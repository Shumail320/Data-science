```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Dataset From Seaborn's Built-in Dataset
df = sns.load_dataset('iris')
```

```python
# Shape of the dataset (rows, columns)
print("Shape:", df.shape)
```

```python
# Column names
```

```python
# First 5 rows
print("\nHead:")
print(df.head())
```

```python
#ScatterPlot(analyze relationship between variables)
```

```python

# List of numeric columns
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Create scatter plots for all unique pairs
for i in range(len(features)):
    for j in range(i + 1, len(features)):
        plt.figure(figsize=(6, 4))
        sns.scatterplot(data=df, x=features[i], y=features[j], hue='species')
        plt.title(f'{features[i].capitalize()} vs {features[j].capitalize()}')
        plt.show()
```

```python
#Histogram(examine data distribution)
```

```python

# List of numeric features
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Loop through each feature
for feature in features:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=feature, kde=True, hue='species', multiple='stack')
    plt.title(f'Distribution of {feature.capitalize()}')
    plt.xlabel(feature.capitalize())
    plt.show()
```

```python
#BoxPlot(detect outliers and spread of values)
```

```python

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for feature in features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, x='species', y=feature)
    plt.title(f'{feature.capitalize()} by Species')
    plt.show()
```
